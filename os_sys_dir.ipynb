```python
try:
    raise IndexError
except:
    print(sys.exc_info())#a tuple of three values that give information about the exception that is currently being handled
```

```python
import traceback
try:
    raise TypeError('already got one')
except:
    print(sys.exc_info())
    exc_info = sys.exc_info()
    print(exc_info[0])
    print(exc_info[1])
    print(exc_info[2])
    traceback.print_tb(exc_info[2])#prints from stack trace
```

### 3. os.walk

```python
import os
for (dirname, subshere, fileshere) in os.walk(r'C:\Windows\Microsoft.NET'):
    print('[' + dirname + ']')
    print(subshere)
    print('-'*30)
    for fname in fileshere:
        print(os.path.join(dirname, fname)) # handle one file
```
